```python
import pandas as pd
df = pd.read_excel('student2.xlsx')
df
```

```python
df.index = ['1번', '2번', '3번', '4번', '5번']
df.index.name = '학번'
df
```

```python
# 데이터 교체(대체)
# 데이터 저장을 위해서는 inplace 속성 추가
df['학급'].replace({'자바반': '수학반'})
df['학급'].replace({'자바반': '수학반', '캐드반' : '과학반'})
```

```python
# 이름 컬럼 반 변경
df['이름'] = 'python'
df
```

```python
# 특정 조건을 만족하는 행과 열을 찾아서 데이터 변경
df.loc[df['학급'] == '캐드반', '학급'] = '과학반'
df
```

```python
# 컬럼 추가 및 해당 컬럼에 데이터 추가

# 총점 컬럼 생성 및 '10' 데이터 추가
df['총점'] = 10
df
```

```python
# 총점 컬럼의 값을 국어, 영어 점수의 합으로 변경
df['총점'] = df['국어점수'] + df['영어점수']
df
```

```python
# '합격여부'라는 컬럼을 생성한 후
# 총점이 160점 이상이면 'pass'
# 160점 미만이면 'fail'이라는 데이터 추가
df['합격여부'] = 'fail'
df.loc[df['총점'] >= 160, '합격여부' == 'pass']
df
```
