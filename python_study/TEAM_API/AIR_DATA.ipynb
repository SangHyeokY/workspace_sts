### 미세먼지 및 이산화탄소량을 통한 '대기질 상태' 데이터
+ 미세먼지, 초미세먼지, 이산화탄소, 공기질
+ AQI 농도

```python
import pandas as pd
```

```python
team_df = pd.read_csv('TeamAPI.csv', encoding='cp949') # 오류나서 encoding 넣음
team_df
```

```python
# 미세먼지
# 좋음 : 0 ~ 30
# 보통 : 31 ~ 80
# 나쁨 : 81 ~ 150
# 매우 나쁨 : 151 ~

# 초미세먼지
# 좋음 : 0 ~ 15
# 보통 : 16 ~ 35
# 나쁨 : 36 ~ 75
# 매우 나쁨 : 76 ~

team_df['미세먼지 등급'] = '보통'
team_df['초미세먼지 등급'] = '보통'
```

```python
# 미세먼지

team_df.loc[(team_df['미세먼지 (㎍/㎥)'] >= 0) & (team_df['미세먼지 (㎍/㎥)'] <= 30), '미세먼지 등급'] = '좋음'
team_df.loc[(team_df['미세먼지 (㎍/㎥)'] > 30) & (team_df['미세먼지 (㎍/㎥)'] <= 80), '미세먼지 등급'] = '보통'
team_df.loc[(team_df['미세먼지 (㎍/㎥)'] > 80) & (team_df['미세먼지 (㎍/㎥)'] <= 150), '미세먼지 등급'] = '나쁨'
team_df.loc[team_df['미세먼지 (㎍/㎥)'] > 150, '미세먼지 등급'] = '매우 나쁨'
team_df
```

```python
# 초미세먼지

team_df.loc[(team_df['초미세먼지 (㎍/㎥)'] >= 0) & (team_df['초미세먼지 (㎍/㎥)'] <= 15), '초미세먼지 등급'] = '좋음'
team_df.loc[(team_df['초미세먼지 (㎍/㎥)'] > 15) & (team_df['초미세먼지 (㎍/㎥)'] <= 35), '초미세먼지 등급'] = '보통'
team_df.loc[(team_df['초미세먼지 (㎍/㎥)'] > 35) & (team_df['초미세먼지 (㎍/㎥)'] <= 75), '초미세먼지 등급'] = '나쁨'
team_df.loc[team_df['초미세먼지 (㎍/㎥)'] > 75, '초미세먼지 등급'] = '매우 나쁨'
team_df
```

###### 대기 중 평균 이산화탄소 415.7ppm / 데이터 자료는 실내 기준
###### 실내 이산화탄소 권장 수치 : 1,000ppm 이하
###### 비고 : 도서관, 영화상영관, 학원, 인터넷컴퓨터게임시설제공업 영업시설 중 자연환기가 불가능하여 자연환기설비 또는 기계환기설비를 이용하는 경우에는 이산화탄소의 기준을1,500ppm 이하로 한다.
###### 실내 기준 : 미세먼지 -> 100(㎍/㎥) 이하, 초미세먼지(㎍/㎥) -> 50 이하, 이산화탄소 -> 1000(ppm) 이하로 권장

```python
# 통합실내지수
# 0 ~ 49 : 매우 나쁨
# 50 ~ 79 : 나쁨
# 80 ~ 89 : 보통
# 90 ~ 100 : 좋음

team_df['통합실내지수 등급'] = '보통'
team_df
```

```python
team_df.loc[(team_df['통합실내지수'] >= 0) & (team_df['통합실내지수'] < 50), '통합실내지수 등급'] = '매우 나쁨'
team_df.loc[(team_df['통합실내지수'] >= 50) & (team_df['통합실내지수'] < 80), '통합실내지수 등급'] = '나쁨'
team_df.loc[(team_df['통합실내지수'] >= 80) & (team_df['통합실내지수'] < 90), '통합실내지수 등급'] = '보통'
team_df.loc[(team_df['통합실내지수'] >= 90) & (team_df['통합실내지수'] <= 100), '통합실내지수 등급'] = '좋음'
team_df
```
